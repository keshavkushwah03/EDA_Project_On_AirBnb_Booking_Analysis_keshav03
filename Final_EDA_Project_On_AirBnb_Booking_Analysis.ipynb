### Variables Description

1. id: A unique identifier for each listing.

2. name: The name or title of the listing.

3. host_id: A unique identifier for the host.

4. host_name: The name of the host.

5. neighbourhood_group: The larger administrative area or borough where the listing is located.

6. neighbourhood: The specific neighborhood or district where the listing is located.

7. latitude: The latitude coordinate of the listing's location.

8. longitude: The longitude coordinate of the listing's location.

9. room_type: The type of room available for booking:-
   - **Private room**- Guests have exclusive access to the bedroom/sleeping area of the listing. Other parts area such as the living room, kitchen, and bathroom are likely open either to the host even to other guests.

   - **Entire home/apt**- Guests have the whole place for themselves. It usually includes a bedroom, bathroom, and kitchen.

   -  **Shared Room**- Guests sleep in a bedroom or a common area that could be shared with others.
    
10. price: The nightly price for booking the listing.

11. minimum_nights: The minimum number of nights required for booking.

12. number_of_reviews: The total number of reviews received for the listing.

13. last_review: The date of the last review for the listing.

14. reviews_per_month: The average number of reviews per month for the listing.

15. calculated_host_listings_count: The number of listings belonging to the same host.

16. availability_365: The number of days the listing is available for booking in a year.

````python
# Plotting Pie chart which gives percentage and total listings of neighbourhood group
labels = copy_airbnb.neighbourhood_group.value_counts().index
colors = ['aqua','teal','red','pink','black']
explode = (0,0,0,0,0)
shape = copy_airbnb.neighbourhood_group.value_counts().values
plt.figure(figsize=(10,15))
plt.pie(shape, explode = explode, labels=shape, colors= colors, autopct = '%1.1f%%', startangle=90, pctdistance=1.1 , labeldistance=1.2)
plt.legend(labels)
plt.title('Neighbourhood Group')
plt.show()
````

````python
# Create a DataFrame to hold the counts of each neighbourhood group
neighbourhood_group_counts = copy_airbnb['neighbourhood_group'].value_counts().reset_index()
neighbourhood_group_counts.columns = ['neighbourhood_group', 'count']

print(neighbourhood_group_counts)

# Create a bar plot
plt.figure(figsize=(8, 5))
plt.bar(neighbourhood_group_counts['neighbourhood_group'], neighbourhood_group_counts['count'], color='m')
plt.title('Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Total Listings')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
````
